```python
df = pd.read_csv("/home/tunde/mygit/thesis_notebooks/confirmed_planets.csv", skiprows=17 )
print(len(df))
df.head(10)
```

```python
options = ['Radial Velocity', 'Transit', 'Transit Timing Variations']

df = df.loc[(df["discoverymethod"].isin(options))]
print(len(df))
```

```python
df
```

```python
np.nanmax(df["pl_rade"])
```

```python
plt.scatter(df["pl_orbper"][np.isnan(df["pl_rade"])]/365, df["pl_bmasse"][np.isnan(df["pl_rade"])],
            marker="*", c= "g", label="Radial Velocity")
plt.xscale("log")
plt.yscale("log")
```

```python
plt.rc('ytick', labelsize=15)
plt.rc('xtick', labelsize=15)
plt.figure(figsize=(12,7))
plt.grid("on", alpha=0.3)
plt.scatter(df["pl_orbper"][np.isnan(df["pl_rade"])], df["pl_bmasse"][np.isnan(df["pl_rade"])],
            marker="*", c= "g", alpha=0.5,label="Radial Velocity")

plt.scatter(df["pl_orbper"], df["pl_bmasse"],c=np.log10(df["pl_rade"]), cmap="plasma",
            alpha=1, label="Transit")
plt.xscale("log")
plt.yscale("log")
cbar = plt.colorbar(ticks=[-0.5,0,0.7,1.3])
cbar.ax.set_yticklabels(np.round(10**cbar.get_ticks(),1))
cbar.set_label('Planet Radius [$R_{\oplus}$]', fontsize=16)
plt.xlim([0.01, 10e4]);
plt.xlabel("Orbital Period [days]", fontsize=16)
plt.ylabel("Planet Mass [$M_{\oplus}$]", fontsize=16);
plt.text(0.01,0.85,"<----Earth",fontsize=12)
plt.text(0.01,17.15,"<----Neptune",fontsize=12)
plt.text(0.01,317.8,"<----Jupiter",fontsize=12)
plt.text(4.7e5,2700,"- $R_{\mathrm{Jup}}$",fontsize=16)

plt.legend();


```
