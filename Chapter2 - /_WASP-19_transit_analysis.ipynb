##### Download TESS data for WASP-19

```python
w19_lc = search_lightcurvefile('WASP-19',mission="TESS",cadence="short",
                              sector=9).download(quality_bitmask="hardest")  
w19_lc
```

```python
#obtain target flux fraction and contamination flux fraction fc
fc = 1 - w19_lc.header(1)["CROWDSAP"]
print(f"contamination flux fraction is fc = {fc:.3f}")
```

```python
lc_temp = w19_lc.SAP_FLUX.remove_nans().remove_outliers(sigma_lower=35,sigma_upper=3)

#decontaminate flux
lc_temp.flux = decontaminate(lc_temp.flux, fc)

#3sigma clipping
_,_,_, pts_clipped = clip_outliers(lc_temp.time, lc_temp.flux, lc_temp.flux_err,
                                3,31, return_clipped_indices=True)
lc = lc_temp[~pts_clipped]
lc_SAP = lc.copy()
lc_clipped_pts = lc_temp[pts_clipped]


ax = lc_clipped_pts.scatter(label="3$\\sigma$ clipped", color="r");
lc.scatter(ax=ax, label="SAP");
```

##### Compute BLS to estimate planet period

```python
pdgm_transit = lc.to_periodogram(method="bls", minimum_period=0.3, maximum_period=15,
                                 duration=np.arange(0.02,0.2,0.001))
```

```python
pmax = pdgm_transit.period_at_max_power.value
tref = pdgm_transit.transit_time_at_max_power
tdur = pdgm_transit.duration_at_max_power.value
print(f"Planet period is {pmax:.4f} days and transit duration is {tdur:.3f} days")
```

##### Compute Lomb Scargle periodogram to estimate period of stellar variability

```python
pdgm_var = lc.to_periodogram(method="ls", minimum_period=0.3, maximum_period=20)
pvar = pdgm_var.period_at_max_power.value
print(f" Stellar variability period is {pvar:.2f} days")
```

##### Plot BLS and LS periodograms

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))

pdgm_transit.plot(ax=ax[0], label="BLS")
ax[0].axvline(pmax, ls="dashed", color="r", alpha=0.5)
ax[0].text(pmax+0.15, 20000, f"$\\longleftarrow${pmax:.3f} days\n      (transit)", c="r", fontsize=14)
ax[0].text(10,15200, f"$t_0$={tref:.2f}\n$t_T$={tdur:.2f}days")


pdgm_var.plot(ax=ax[1], label="Lomb_scargle")
ax[1].axvline(pvar, ls="dashed", color="r", alpha=0.5)
plt.text(pvar+0.15, 20, f"$\\longleftarrow${pvar:.2f} days\n       (star)", c="r", fontsize=14)


plt.tight_layout()
```

### Model stellar variability

```python
## convert to relative flux units
lc= lc.normalize()
```

```python
oot = pdgm_transit.get_transit_mask(pmax,tdur*1.3,tref)
ax = lc[oot].scatter(label="OOT points")
lc[~oot].scatter(ax=ax, color="r", label="transit points");
```

### Detrend the light curve

##### Savitzky-Golay filter

```python
lc_SG, SG_trend = lc.flatten(501, polyorder=3, mask=~oot, return_trend=True)
```

```python
ax = lc.scatter(label="data")
SG_trend.plot(ax=ax, color="g",lw=3, label="Savitzky Golay");
```

#### Gaussian processes

```python
import celerite
from celerite import terms

ppm=1e-6
m32 = terms.Matern32Term(log_sigma = np.log(1000*ppm), log_rho = np.log(4),
                         bounds = dict(log_sigma=(-15, 10), log_rho=(-15, 10)) )

jt = terms.JitterTerm(log_sigma = np.log(np.std(lc.flux)), 
                  bounds = dict(log_sigma = (-15,5)))

kernel = m32 + jt

gp = celerite.GP(kernel, mean=1)
gp.compute(lc[oot].time, lc[oot].flux_err)  # You always need to call compute once.

bounds = gp.get_parameter_bounds()

```

```python
from scipy.optimize import minimize, differential_evolution

def neg_log_like(params, f, gp):
    gp.set_parameter_vector(params)
    return -gp.log_likelihood(f)

r = differential_evolution(neg_log_like, args=(lc[oot].flux, gp), bounds=bounds,polish=True)
r.x
```

```python
gp.set_parameter_vector(r.x)
gp_mean = gp.predict(lc[oot].flux, lc.time,return_cov=False, return_var=False)

ax = lc.scatter()
SG_trend.plot(ax=ax, color="g",lw=3, label="Savitzky Golay");
plt.plot(lc.time, gp_mean, "r");
```

#### Divide the data by the GP model

```python
det_lc = lc.copy()
det_lc.flux, det_lc.flux_err = det_lc.flux/gp_mean, det_lc.flux_err/gp_mean
```

```python
det_lc.scatter(label="detrended light-curve");
```

### Transit fitting using Differential Evolution

```python
import batman
def bat_model(pars, time):
    params = batman.TransitParams()

    params.rp = pars[0]                              #planet radius (in units of stellar radii)
    params.a = pars[1]                               #semi-major axis (in units of stellar radii)
    imp_par = pars[2]
    params.inc = inclination(imp_par, params.a)      #orbital inclination (in degrees)
    params.ecc = 0                                   #eccentricity
    params.w = 90.                                   #longitude of periastron (in degrees)
    u1, u2 = convert_LD_coeffs(pars[3], pars[4],"q2u",verify=False)       #convert kipping
    params.per = pars[5]
    params.t0 = pars[6]                            #time of inferior conjunction


    params.u = [u1, u2]                              #limb darkening coefficients
    params.limb_dark = "quadratic"                   #limb darkening model
    m = batman.TransitModel(params, time)
    batflux = m.light_curve(params)
#     print(m.fac)
          
    return batflux#*pars[7]
```

```python
det_lc.time = np.array([t for t in det_lc.time])
```

```python
def neg_ll(pars):
    chi2 = np.sum( np.square( (det_lc.flux - bat_model(pars, det_lc.time))/det_lc.flux_err  ))
    return chi2


bounds = ( (0.05,0.18), (3,4), (0,1), (0,1), (0,1),(pmax-0.05,pmax+0.05),(tref-0.2,tref+0.2))
result = differential_evolution(neg_ll, bounds=bounds,polish=True)
result.x
```

```python
print(f"The minimum chi_squared value is {result.fun:.2f}")
```

### Overplot transit model on the data

```python
det_lc.scatter(label="data")
plt.plot(det_lc.time, bat_model(result.x, det_lc.time), "r", label = "best-fit model")
plt.legend(fontsize=12)
```

```python
mle_p  =  result.x[-2]
mle_t0 = result.x[-1]
print(f"Maximum Likelihood Estimate for $P$ and $t_0$ are: {mle_p:.7f} and {mle_t0:.7f}")
```

### Phase fold

```python
# %matplotlib inline

lc_fold = det_lc.fold(mle_p,mle_t0)
phase_pars = result.x.copy()
phase_pars[-2:] = [1, 0]
mle_model = bat_model(phase_pars, lc_fold.phase)
```

```python
lc_fold.scatter(label="phase-folded data")
plt.plot(lc_fold.phase, mle_model, "r", label= "transit model")
plt.legend(fontsize=12)
plt.xlim([-0.15,0.15]);
```
