```python
def model(sim_pars, phase, LDC=True, ld_law="linear"):
    params = batman.TransitParams()
    
    params.limb_dark = ld_law                         #limb darkening law
    params.ecc = 0                                    #eccentricity
    params.w = 90                                     #longitude of periastron (in degrees)
    
    params.per = 1                                    #orbital period
    params.rp = sim_pars[0]                           #planet radius (in units of stellar radii)
    params.a = sim_pars[1]                            #semi-major axis (in units of stellar radii)
    imp_par = np.abs(sim_pars[2])
    params.inc = inclination(imp_par, params.a)#      #orbital inclination (in degrees)
    params.t0 = sim_pars[3]                           #time of inferior conjunction
    
    if LDC: 
        u = np.abs(sim_pars[4]) if ld_law is "linear" else convert_LD_coeffs(np.abs(sim_pars[4]),
                                                                             np.abs(sim_pars[5]), "q2u",verify=False)
    else: 
        u = 0 if ld_law is "linear" else [0, 0]
    
    params.u = [u] if ld_law is "linear" else u                      #limb darkening coefficients
    m = batman.TransitModel(params,phase)
    
    return m.light_curve(params)
```

#### Planet parameters: Jupiter-sized planet with 10-day orbital Period and impact parameter of 0.5

```python
#parameters of the planet transit
P = 10
Rp = 0.1
aR = P_to_aR(P,1,1)
b=0.5
t0=0

time = np.linspace(-3, 3, int(6*60/0.5))   #in hours. cadence of 0.5min
time_15m = np.linspace(-3, 3, int(6*60/15))   #in hours. cadence of 15min

phases = time/(24*P)
phases_15m = time_15m/(24*P)
```

```python
sim_pars = [Rp, aR, b, t0, uln]
print(sim_pars)
```

#### Simulate transit models for limb-darkened and non limb-darkened star

```python
modelsim_LD = model(sim_pars, phases, ld_law="linear")
modelsim_LD_15m = model(sim_pars, phases_15m,ld_law="linear")

modelsim_nLD = model(sim_pars, phases, False, ld_law="linear")
```

```python
fig,ax = plt.subplots(2,1, figsize=(9,6), gridspec_kw={"height_ratios":(3,1)}, sharex=True)
ax[0].plot(phases, modelsim_LD, "k", lw=2 , label = "Model with LD")
ax[0].plot(phases, modelsim_nLD, "r", lw=2, alpha=0.5,label = "Model without LD")
ax[0].set_ylabel("Flux");
ax[0].legend(loc="upper center")

resid = (modelsim_LD - modelsim_nLD)*1e6
amp = np.max(abs(resid))

ax[1].axhline(0,ls="dashed", alpha=0.5)
ax[1].plot(phases, resid, "r.-", lw=0.5, label = f"amp={amp:.2f}ppm")


ax[1].legend(loc="upper center", fontsize=14)
ax[1].set_xlabel("Phases")
ax[1].set_ylabel("diff [ppm]")
plt.subplots_adjust(hspace=0.01)
plt.xlim([min(phases), max(phases)]);

```

The signature of LD is not the difference between the model with and without LD but actually a non-LD fit to the model with LD

### obtain the observable signature of limb darkening from fitting with non-LD model

```python
from scipy.optimize import minimize, differential_evolution

def min_f(sim_pars, ph, data):
    return np.sum((model(sim_pars, ph, False) - data)**2)

# r = minimize(min_f, x0=np.array([Rp,aR,b,t0])*1.01, args=(phases, modelsim_LD))
bounds= [(0.5*Rp, 1.5*Rp), (0.5*aR, 1.5*aR), (b-0.2, b+0.4), (-0.001, 0.001) ]

r = differential_evolution(min_f, args=(phases, modelsim_LD), bounds = bounds,
                           polish=True,seed=1153, workers=-1)
print(f"The parameters, [Rp, a/R*, b, t0], of the non_LD model fit to the simulated LD transit are:\n{r.x}")
```

```python
#fit of the simulation with 15m cadence
r_15m = differential_evolution(min_f, args=(phases_15m, modelsim_LD_15m),bounds = bounds, polish=True )
r_15m.x
```

Notice how the paramters of the nLD model fit has been slightly modified to mimic the LD transit.

```python
nLD_fit = model(r.x, phases, False) 
nLD_fit_15m = model(r_15m.x, phases_15m, False) 

tdur = transit_duration(P,Rp,aR,b=b)/(24*P)
ing_dur = ingress_duration(P,Rp,aR,b=b)/(24*P)

tdur_nLD = transit_duration(P,r.x[0],r.x[1],b=r.x[2])/(24*P)
ing_dur_nLD = ingress_duration(P,r.x[0],r.x[1],b=r.x[2])/(24*P)

fig,ax = plt.subplots(2,1, figsize=(9,6), gridspec_kw={"height_ratios":(3,1)}, sharex=True)
ax[0].plot(phases, modelsim_LD, "k", lw=2 , label = "Simulated transit with LD")
ax[0].plot(phases, nLD_fit, "r--", lw=2, label = "non-LD fit")
ax[0].set_ylabel("Flux");
ax[0].legend(loc="upper center")
# ax[0].axvspan(-tdur/2, -tdur/2+ing_dur,alpha=0.5, color="k" )


resid = (modelsim_LD - nLD_fit)*1e6
resid_15m = (modelsim_LD_15m - nLD_fit_15m)*1e6

amp = np.max(abs(resid))
amp_15m = np.max(abs(resid_15m))

ax[1].axhline(0,ls="dashed", alpha=0.5)
ax[1].plot(phases, resid, "r.-", lw=0.5, label = f"amp={amp:.2f}ppm")
ax[1].plot(phases_15m, resid_15m, "bo-", lw=0.5)
ax[1].axvspan(-tdur_nLD/2, r.x[-1]-tdur_nLD/2+ing_dur_nLD,alpha=0.1, color="r" )
ax[1].axvspan(tdur_nLD/2, r.x[-1]+tdur_nLD/2-ing_dur_nLD,alpha=0.1, color="r" )



ax[1].legend(loc="upper center", fontsize=14)
ax[1].set_xlabel("Phases")
ax[1].set_ylabel("Res[ppm]")
plt.subplots_adjust(hspace=0.01)
plt.xlim([min(phases), max(phases)]);

```

## Detect the limb darkening feature in simulated observation

One could be faced with the problem of determining if a particular observation has LD effects included or not and this might not be clear-cut with the noise-level of the observation.

Let's imaging a limb darkened transit observation with noise-level almost twice the amplitude of the LD signature

```python
# 8911, 5289, 7621
#2074 linear 9525(b12=31)
seed_value = 2074#np.random.random_integers(0,10000)
```

```python
nl = 2*amp * 1e-6        #noise level
print(f"seed is {seed_value}")
np.random.seed(seed_value)
noise = np.random.normal(1, nl, len(phases))

flux_obs = modelsim_LD*noise    #simulated limb darkened flux with noise
err = np.ones_like(noise)* nl
```

### show the simulated model

```python
plt.figure(figsize=(9,5))
plt.errorbar(phases, flux_obs, err, fmt="k.", ms=10, ecolor="gray", alpha=0.2,zorder=0)
plt.plot(phases, modelsim_LD, "b", lw=2 , label = "Simulated transit with LD")
plt.plot(phases, modelsim_nLD, "r", lw=2, alpha=0.7,label = "Model without LD")
plt.legend()
plt.xlabel("Phases")
plt.xlim([min(phases), max(phases)]);
plt.ylabel("Flux");
```

Visually, both models seem to match the data well. We can perform model comparison to test which model is most probable given this transit observation

```python
#likelihood function
def lnlike(params, LDC=True):    
    tmodel = model(params,phases, LDC)
    sigma2 = err ** 2
    return -0.5 * np.sum((flux_obs - tmodel) ** 2 / sigma2 + np.log(sigma2))   
```
