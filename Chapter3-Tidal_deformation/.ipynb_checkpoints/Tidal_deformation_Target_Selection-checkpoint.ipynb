```python
def shape(Rv,aR,hf,qmass):
    qr = 0.5*hf*1/qmass*(Rv/aR)**3
    bx = Rv*(1-(2./3.)*qr + (17./9.)*qr**2 - (328./81.)*qr**3 + (2558./243.)*qr**4)
    ax = bx*(1.+3.*qr)          #eqn(10) correia 2014
    cx = bx*(1.-qr)
    abc=np.array([ax,bx,cx])
    return abc
```

```python
#filter data
pls = planets_df.loc[(planets_df["M_p[M_Jup]"]>0) & 
                     (planets_df["R_p[R_Jup]"]>0.2) & 
                     (planets_df["P"]>0) &
                     ( (np.isfinite(planets_df["i_p"])) | (np.isfinite(planets_df["b"])) )]

w103_ind = pls.index[pls["planet"] == "WASP-103b"].tolist()
print(w103_ind)

# update parameters of w103 to values from southworth&evans 2016
pls.loc[w103_ind, ["Rst","Mst","i_p","b","R_p[R_Jup]",
                  "M_p[M_Jup]","a[AU]","pl_refname"]] = 1.413, 1.205, 88.2, np.nan, 1.596, 1.47, 0.01979, "Southworth_et_al_2016 href=https://ui.adsabs.harvard.edu/abs/2016MNRAS.463...37S/abstract"

# pls.loc[w103_ind, ["Rst","Mst","i_p","b","R_p[R_Jup]",
#                   "M_p[M_Jup]","a[AU]","pl_refname"]] = 1.416, 1.21, 88.8,0.06, 1.681, 1.51, 0.01979, "Delrez_et_al_2016 "
```

```python
#radius ratio and mass ratio
pls = pls.assign(Rp_Rs = (pls["R_p[R_Jup]"]*c.R_jup.value/(pls["Rst"]*c.R_sun.value)),
                 Mp_Ms = (pls["M_p[M_Jup]"]*c.M_jup.value/(pls["Mst"]*c.M_sun.value)))

pls = pls.assign( Rr=2.46*((1/pls["Mp_Ms"])**(1/3.))*pls["Rp_Rs"],
                 aR = pt.convert_param.AU_to_aR(pls["a[AU]"],pls["Rst"]))

nan_ind_b = pls.index[np.isnan(pls["b"])].tolist()
for ind in nan_ind_b:
    pls.at[ind, "b"]=  pls["aR"][ind]*np.cos(np.deg2rad(pls["i_p"][ind]))
    
pls = pls.loc[(pls["b"])<0.8]
#distance to star in roche units
pls = pls.assign(R0 = pls["aR"]/pls["Rr"])

#asymmetry parameter between axis of triaxial ellipsoid assuming hf=1.5
hf=1.5
pls = pls.assign(q = hf*0.5/pls["Mp_Ms"]*(pls["Rp_Rs"]/pls["aR"])**3)

#select planets with the  largest asymmetry q (most deformed) and sort
def_pls = pls.loc[(pls["q"]>0.01)]
def_pls = def_pls.drop(columns=['a[AU]', 'pl_ratror'])
def_pls["$\Delta$R"] = def_pls["Rp_Rs"] * 7/3*def_pls["q"]#def_pls["def%"]/100 * def_pls["Rp_Rs"]**2 *(1-def_pls["b"]**2)**1.42

def_pls["def [%]"] = 100* def_pls["$\Delta$R"] /def_pls["Rp_Rs"]
def_pls = def_pls.loc[(def_pls["def [%]"] > 5)]

def_pls = def_pls.sort_values(by="q", ascending=False)
```

```python
P = def_pls.at[w103_ind,'P']                                    #days
t0 = 0                                             #mid transit time
aR = def_pls.at[w103_ind,'aR']                                         #scaled semimajor axis a/R*
Rv = def_pls.at[w103_ind,'Rp_Rs']                                         # planet radius ratio
inc = def_pls.at[w103_ind,'i_p']                                         # orbital inclination
b = pt.convert_param.impact_parameter(inc,aR)

u1, u2 = 0.5343,0.1299
ldc_1 = [u1,u2]                                    #limb darkening parameters 

qmass = def_pls.at[w103_ind,'Mp_Ms']                                     #ratio of planet to stellar mass

phases=np.linspace(-0.1,0.1,300)  
```

```python
#set shape of the planet shape_2 = "love" and use the following values for hf:
hf_values=[2.5,1.5,1, 0]
plot_phases = np.linspace(-0.5,0.5,100)

col = ["m","y","b","c"]
#for ellc
r_1 = 1/aR    #radius of star in units of the semimajor axes
r_2 = Rv/aR   #radius of the planet in units of the semimajor axis

ellc_sphere = ellc.lc(phases, t_zero=t0, radius_1=r_1, radius_2=r_2, incl=inc, sbratio=0,
                      ld_1="quad", ldc_1=ldc_1, shape_1='sphere', shape_2='sphere',
                      grid_1="very_fine", grid_2="very_fine") 


fig, ax = plt.subplots(2,1,figsize=(10,12))
ax[0].set_title("Light-curves of a tidally deformed planet at different values  of $h_f$", fontsize=15)
ax[1].set_title("Projected Area of ellipsoidal with phase", fontsize=15)


for i,h_f in enumerate(hf_values):
    ellc_flux = ellc.lc(phases,t_zero=t0, radius_1=r_1,radius_2=r_2,incl=inc,sbratio=0,
                      ld_1="quad",ldc_1=ldc_1,shape_1='sphere',shape_2='love',
                      q=qmass, hf_2= h_f, grid_1="fine", grid_2="fine") 
    ax[0].plot(phases, ellc_flux, col[i], label = "Ellipsoidal: $h_f$ = {:.1f}".format(h_f), lw=2)
    
    ellps_area, abc = ellc.projected_ellipse_area(plot_phases, Rv, aR, h_f, qmass, inc)
    
    ax[1].plot(plot_phases,ellps_area/(np.pi*Rv**2), col[i], label = "Ellipsoidal: $h_f$ = {:.1f}".format(h_f), lw=2)

    

#overplot spherical planet light curve    
ax[0].plot(phases, ellc_sphere, "r--",label="Spherical", lw=2,alpha=0.6)
ax[1].plot(plot_phases, np.ones_like(plot_phases), "r--",label="Spherical", lw=3,alpha=0.6)


ax[0].text(-0.097,0.998, "(a)", fontsize=18)
ax[0].set_ylabel("Flux",fontsize=18)
ax[0].legend(fontsize=15)

ax[1].set_ylabel("Normalized Projected Planet Area")
ax[1].text(-0.097,1.01, "(b)", fontsize=18)


# ax[1].legend(fontsize=14)

plt.xlabel("Phase", fontsize=18);
plt.subplots_adjust(wspace=0.15)
[ax[i].set_xlim([min(phases), max(phases)]) for i in range(2)]
ax[1].set_ylim([0.85, 1.02])
[ax[i].set_xticks([-0.1,-0.05,0,0.05,0.1]) for i in range(2)];
ax[1].set_yticks(np.arange(0.84,1,0.04));
ax[0].set_yticks(np.arange(1,0.982,-0.004));
```

Notice that  the  light-curve  of  the  ellipsoidal model  changes  noticeably  for  different  values  of $h_f$ and  also compared to the spherical case. This is because the ellipsoidal planet projects only a small cross-section of its shape during the transit thereby leading to a lower transit depth when compared to the spherical planet. A higher $h_f$ for the same planet causes more deformation, which leads to even smaller projected cross-sectional  area and thus a shallower tansit depth.

The case  where $h_f$ = 0 (although not physical) implies no deformation to the planet so that the ellipsoidal planet model is equivalent to that of a spherical planet (red curve and dashed cyan curve)

### Spherical model fit of simulated transit of deformed WASP-103 (assuming it is oblate and with $h_f=1.5$)

```python
from Planet_tools.convert_param import inclination, impact_parameter, convert_LD_coeffs

def model(params,t,pl_shape="sphere",qmass=qmass,data=None, ldc=[u1,u2]):
    #params[a,rp,b,t0,q1,q2]
    
    r_1 = 1./params[0]                          #Rst/a
    r_2 = params[1]/params[0]                   # Rp/a
    incl = inclination(params[2],params[0])     #inclination 
    #u1_,u2_= params[3],params[4]    
    ldc_1 = ldc
    ellc_flux = ellc.lc(t, t_zero=t0, radius_1=r_1,radius_2=r_2,incl=incl,sbratio=0,ld_1="quad",
                    ldc_1=ldc_1,shape_1='sphere',shape_2=pl_shape, grid_1='default',
                        grid_2='default',q=qmass,hf_2=1.5)
    if data is None:
        return ellc_flux
    return np.sum((data - ellc_flux)**2) 
```

```python
params = [aR, Rv, b]

flux_def = model(params,phases,"love", qmass)     #simulate deformed

plt.plot(phases, flux_def);
```

```python
#fit spherical planet transit to the deformed planet flux(flux_def)
from scipy.optimize import minimize, least_squares, differential_evolution

result = differential_evolution( model, args=(phases, "sphere",qmass, flux_def), 
                                bounds=((2.3,3.5),(0.08,0.16),(0.,0.5)),
                                maxiter=500,popsize=5,polish=True)
print(result.x)
```

```python
#transit width of ellipsoidal planet and spherical planet fit
aR_sphr, Rp_sphr, b_sphr= result.x                          #, u1_sphr, u2_sphr 
inc_sphr =pt.convert_param.inclination(b_sphr, aR_sphr)

W_ellps14 = tdur(P,Rv,aR,inc=inc)/(P*24)
W_ellps23 = tdur(P,Rv,aR,inc=inc, total=False)/(P*24)
```

```python
spr_fit = model(result.x,phases,"sphere")

fig,ax = plt.subplots(2,1,figsize=(12,10), gridspec_kw={"height_ratios":(3,1)}, sharex=True)
ax[0].set_title("Spherical model fit to deformed WASP-103",fontsize=15)
ax[0].plot(phases, flux_def, ".",
           label = f"simulated deformed: $R_{{\mathrm{{v}}}}$={Rv:.3f}$R_{{\\ast}}$, ${{a/R_{{\\ast}}}}$={aR:.2f}, b={b:.2f}, $h_f$ = 1.5")
ax[0].plot(phases,spr_fit, "r-",
          label = f"Spherical fit: $R_{{\mathrm{{spr}}}}$={Rp_sphr:.3f}$R_{{\\ast}}$, ${{a/R_{{\\ast}}}}$={aR_sphr:.2f}, b={b_sphr:.2f}")
 
ax[0].legend()
ax[0].set_ylim([None, 1.005])
ax[0].set_yticks(np.arange(0.985,1.005,0.005))
ax[0].set_ylabel("Flux", fontsize=18)


ax[1].axhline(c="k",ls="dashed")
ax[1].plot(phases, 1e6*(flux_def-spr_fit), "r", label="amp={:.2f} ppm".format(np.max(np.abs(1e6*(flux_def-spr_fit)))) )
ax[1].legend(fontsize=15)
ax[1].set_xlabel("Phase", fontsize=18)
ax[1].set_ylabel("Residual [ppm]");
ax[1].set_xticks([-0.1,-0.05,0,0.05,0.1])

plt.subplots_adjust(hspace=0.01)

[ax[i].autoscale(enable=True, axis='x', tight=True) for i in range(2)]

plt.axvspan(-W_ellps14/2,-W_ellps23/2,color="r", alpha=0.1);
plt.axvspan(W_ellps14/2,W_ellps23/2,color="r", alpha=0.1);
```

### detectability of tidal deformation

```python
from scipy.stats import uniform, norm

def fit_elps(params,t):
    #params[a,rp,b,t0,q1,q2]
    
    r_1 = 1./params[0]                          #Rst/a
    r_2 = params[1]/params[0]                   # Rp/a
    incl = inclination(params[2],params[0])     #inclination 
    hf = params[3]
    u1_,u2_= params[4],params[5]    
    ldc_1 = [u1_,u2_]
    ellc_flux = ellc.lc(t, t_zero=t0, radius_1=r_1,radius_2=r_2,incl=incl,sbratio=0,ld_1="quad",
                    ldc_1=ldc_1,shape_1='sphere',shape_2="love", grid_1='default',
                        grid_2='default',q=qmass,hf_2=hf)
    return ellc_flux
```

```python
params = [aR, Rv, b, 1.5,u1,u2]

phases= np.linspace(-2.5/24, 2.5/24, int(5*60))/def_pls["P"][w103_ind]
flux_def = fit_elps(params,phases)     #simulate deformed

noise50 = np.random.normal(1, 50*1e-6,len(phases))
flux_obs = flux_def*noise50

plt.plot(phases, flux_obs,".");
```

```python
labels = ["$a/R_{\\ast}$", "$R_{\mathrm{v}}$", "$b$", "$h_{f}$", "$u_1$", "$u_2$"]
ndim, nwalkers = len(labels), 64

          
          
priors = [buni(2.5,3.5), buni(0.05,0.15), buni(0,0.5), buni(0,2.5), norm(u1,0.0012), norm(u2,0.0027)]


def lnlike(params, t, y, yerr):
    return -0.5 * np.sum(((y - fit_elps(params, t))/yerr) ** 2)

  
def lnprior(params):
    lp = 0.0
    for par, pr in zip(params, priors):
        lp += pr.logpdf(par)
    return lp

def prior_transform(u):
    x = np.array(u)
    
    for i, pr in enumerate(priors):
        x[i] = pr.ppf(u[i])
    return x
        

def lnprob(params, t, y, yerr):
    lp = lnprior(params)
    return lp + lnlike(params, t, y, yerr) if np.isfinite(lp) else -np.inf
```
