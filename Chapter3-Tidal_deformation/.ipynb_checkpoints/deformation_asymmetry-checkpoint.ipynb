```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('ytick', labelsize=15)
plt.rc('xtick', labelsize=15)
plt.rc('font', size=18)
```

The asymmetry parameter $q$ of an ellipsoid is given by:

$    q=\frac{h_{f}}{2}\frac{M_{\ast}}{M_{p}}\left(\frac{R_{v}}{d}\right)^3 $

The Roche radius of the star is given by:

$R_{Roche}= 2.46\,R_{v} \left( \frac{M_{\ast}}{M_{p}} \right ) ^{1/3}$

For a planet orbiting at some multiple (n) of roche radius of the star d = $n \times R_{Roche}$

so that $q$ becomes:
    
$q = \frac{h_f}{2} *({2.46\,n})^{-3}$ 

We can then show how the planet deformation reduces with distance from the star from $1R_{roche}$ to $6R_{Roche}$ for different values of $h_f$

```python
n = np.array([1, 1.2, 1.5, 2, 2.5, 3, 3.5, 4.5, 6])
```

```python
q = lambda hf, n: hf/2* 1/(n*2.46)**3
```

```python
fig=plt.figure(figsize=(7,5))
plt.plot(n, q(2.5, n),"ro-",label="$h_f = 2.5$")
plt.plot(n, q(1.5, n),"bo-",label="$h_f = 1.5$")
plt.axhline(0, alpha=0.5, c="k")
plt.xlabel("semi-major axis [stellar radii]")
plt.ylabel("q")
plt.legend();
```

```python
print(f"The maximum value of q is {max(q(2.5,n)):.3f}")
```
